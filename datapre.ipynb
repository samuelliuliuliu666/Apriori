```python
import pandas as pd

# Read CSV file
df = pd.read_csv('NYPD_Hate_Crimes.csv')

# show 10 rows of the dataframe
df.head(10)
```

```python
columns_to_drop = ['Full Complaint ID', 'Complaint Year Number', 'Month Number','Record Create Date','Complaint Precinct Code','Arrest Date','Arrest Id']
df = df.drop(columns=columns_to_drop)
df.head(10)
```

```python
#show the shape
df.shape
```

```python
#drop rows contain missing values if exists
df = df.dropna()
df.shape
```

```python
#output the preprocessed data as INTEGRATED-DATASET.csv
df.to_csv('INTEGRATED-DATASET.csv', index=False)
```
